```python
import requests
import pandas as pd
import os
import json as js

# Ensure your API key is set in the environment variables
foursquare_api_key = os.getenv('FOURSQUARE_API_KEY')
if not foursquare_api_key:
    raise ValueError("FOURSQUARE_API_KEY environment variable not set")

# Foursquare API endpoint and headers
url = "https://api.foursquare.com/v3/places/search"
headers = {
    "Accept": "application/json",
    "Authorization": foursquare_api_key
}

# Function to get venues for a given latitude and longitude
def get_venues(latitude, longitude, radius=1000):
    params = {
        "ll": f"{latitude},{longitude}",
        "radius": radius
    }
    # Send a GET request
    response = requests.get(url, headers=headers, params=params)
    
    # Check if the request was successful
    if response.status_code == 200:
        print(f"Request was successful for {latitude}, {longitude}!")
        # Parse the response content
        data = response.json()
        return data
    else:
        print(f"Request failed with status code: {response.status_code} for {latitude}, {longitude}")
        return None

# Iterate through each bike station and get nearby venues
foursquare_results = []
for index, row in df.iterrows():
    latitude = row['latitude']
    longitude = row['longitude']
    venues_data = get_venues(latitude, longitude)
    if venues_data:
        venues = venues_data.get('results', [])
        for venue in venues:
            venue_info = {
                "station_name": row['name'],
                "latitude": latitude,
                "longitude": longitude,
                "venue_id": venue.get('fsq_id'),
                "venue_name": venue.get('name'),
                "venue_category": venue['categories'][0]['name'] if venue.get('categories') else None,
                "venue_address": venue['location'].get('address') if venue.get('location') else None,
                "venue_distance": venue.get('distance'),
                "venue_latitude": venue['geocodes']['main']['latitude'] if venue.get('geocodes') else None,
                "venue_longitude": venue['geocodes']['main']['longitude'] if venue.get('geocodes') else None
            }
            foursquare_results.append(venue_info)

# Convert results to a DataFrame
foursquare_df = pd.DataFrame(foursquare_results)


```

```python
foursquare_df.to_csv('four_square.csv', index=False)
```
